<h2> Transoformada Z </h2>
<p> Se escogió el polinomio $ y[n] = 5,4y[n-2]-2y[n-2]+ x[n-1]+ 5x[n-1]$ <br>
que al aplicarle la transformada Z nos dará:
$ Y[z] = 5,4z^{-2}Y[z] - 2z^{-2}Y[z] + z^{-1}X[z] + 5z^-1X[z]$
<br>
Al despejar para encontrar la función de tranferencia nos da:
$H(z) = \frac{6z}{z^2-3,4}$

```python
import ztrans as zt
num = [6, 0]
den = [1, 0, -3.4]
zt.mfreqz(num, den)
zt.impz(num, den)
zt.zeropoles(num, den)
```

<h3> Ubicación de polos y ceros </h3>
<p> <strong> Polos: </strong> están en -2 y 2 <br> <strong> Ceros: </strong> está en 0.</p>
<h3> Respuesta en frecuencia y fase </h3><p>Se nota que en amplitud tiende a tener un comportamiento oscilatorio, si se llega a ampliar más la ventana se ve un comportamiento oscilatorio, también se ve que su amplitud está confinado entre 3 y 8dB. En cuanto a la fase se puede observar que va aumentando con un punto de inflexión en 0. Cambia de fase, aumenta al aumentar la frecuencia.<br><p>
<h3> Respuesta al impulso y al escalón </h3>
<p>La respuesta al impulso es un impulso también en sí mientras la respuesta al escalón son dos impulsos separados solamente 1 unidad. La amplitud corresponde a 6 ya que este es el valor de la ganancia (El valor que está en el numerador) <p><br>
<h3> Estabilidad </h3>
Inestable ya que los polos están por fuera de la circunferencia unitaria y también este hecho se puede comprobar en la respuesta de amplitud en frecuencia por su comportamiento oscilatorio <p> 


```python
import numpy as np
fs = 8000
n_1 = np.arange(0, 50, 1/fs)
freq = 10*n_1**2
signal_audio = np.sin(2*np.pi*freq*n_1)
from IPython.display import Audio
Audio(signal_audio, rate=fs)
```

<p> Se puede escuchar que el audio tiene una frecuencia variable, ya que va cambiando el sonido desde más grueso a más agudo y desde más agudo hasta más grueso <p>

```python
import scipy.signal as signal
n = 10
num = signal.firwin(n, [0.2, 0.4], pass_zero = False)
den = np.zeros(n)
den[0] = 1
zt.mfreqz(num, den)
zt.impz(num, den)
zt.zeropoles(num, den)
```

<p> Se puede concluir que deja pasar frecuencias entre 0 y 2.5 Hz disminuyendo a medida que va aumentando la frecuencia. Invierte la frecuencia hasta 2.5 Hz para luego reseatearla. Se puede observar que el filtro no es estable por el lugar de los ceros (por fuera de la circunferencia unitaria), además que se nota un comportamiento periodico pero amortiguado en la respuesta al tiempo <p>

```python
import matplotlib.pyplot as plt
plt.plot(n_1[0:50000], signal_audio[0:50000])
plt.show()
respuesta = np.convolve(signal_audio, num, mode='same')
plt.plot(n_1[0:50000], respuesta[0:50000])
plt.show()
```

<p> Se puede observar que hay cortes en la frecuencia principalmente en las frecuencias más bajas, en las más altas hay un menor corte </p>

```python
lista = [20, 120, 199]
for n in lista:
    num = signal.firwin(n, [0.2, 0.4], pass_zero = False)
    den = np.zeros(n)
    den[0] = 1
    zt.mfreqz(num, den)
    zt.impz(num, den)
    zt.zeropoles(num, den)
    plt.plot(n_1[0:50000], signal_audio[0:50000])
    plt.show()
    respuesta = np.convolve(signal_audio, num, mode='same')
    plt.plot(n_1[0:500000], respuesta[0:500000])
    plt.show()
```

<p> Se puede observar que en la respuesta en freecuencia cada vez se van filtrando menos frecuencias, hay más lóbulos que se dejan pasar y las frecuencias centrales, las de menor valor son cada vez filtradas menos, se tiende a aplanar este lóbulo y a cubrir muchas más frecuencias sin ninguna atenuación. Mientras que las frecuencias mayores van siendo filtradas cada vez más, llegan con menos energía. Se logra ver en el diagrama de polos y ceros que los ceros tienden a organizarse al rededor de la circunferencia unitaria. Menos dos, uno que está en el interior de la circunferencia y otro que está muy alejado de esta, dándole inestabilidad. </p>

```python
fs = 8000
n_1 = np.arange(0, 50, 1/fs)
freq = n_1**3
signal_audio = np.sin(2*np.pi*freq*n_1)
from IPython.display import Audio
Audio(signal_audio, rate=fs)
```

```python
n = 10
num = signal.firwin(n, [0.2, 0.4], pass_zero = False)
den = np.zeros(n)
den[0] = 1
zt.mfreqz(num, den)
zt.impz(num, den)
zt.zeropoles(num, den)
plt.plot(n_1[0:50000], signal_audio[0:50000])
plt.show()
respuesta = np.convolve(signal_audio, num, mode='same')
plt.plot(n_1[0:50000], respuesta[0:50000])
plt.show()
```

<p> Se puede observar un comportamiento casi idéntico al de la señal con la feecuencia varaida cuadrácticamente </p>

```python
n = 10
num = signal.firwin(n, [0.2, 0.4], pass_zero = False)
den = np.zeros(n)
den[0] = 1
zt.computeZ(num, den, [complex(0.5, 0.748), complex(0.5, -0.748)], [1.2], signal_audio)
```

<p> El sistema se ha desestabilizado por el polo, la respuesta en feecuencia del filtro cambió, siendo diferente en cuanto al lóbulo principal, ya que este ya no es el que contiene la mayor amplitud en dbs sino que lo es el lóbulo anterior, en frecuencias menores, corriendo el centro y la forma del filtro <p>
